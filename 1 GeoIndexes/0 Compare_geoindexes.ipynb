Explanation of what is a geoindex. Then, what is GeoHash, H3, S2. Pros and Cons. Which one fits our purpose the most? What size are we picking?

### GeoHash 

https://en.wikipedia.org/wiki/Geohash
https://www.elastic.co/guide/en/elasticsearch/reference/current/search-aggregations-bucket-geohashgrid-aggregation.html

```python
from polygon_geohasher.polygon_geohasher import polygon_to_geohashes, geohashes_to_polygon, geohash_to_polygon
from shapely import geometry
```

```python
def get_geohash_from_bounding_box(gdf):
    HASH_SIZE = 7 #geohash length
    hash_col = 'hash' + str(HASH_SIZE)
    
    gdf_geohash_list = polygon_to_geohashes(gdf.geometry[0], HASH_SIZE, False)
    
    # create an empty dataframe with all the hex
    all_geohash_df = pd.DataFrame(gdf_geohash_list)
    all_geohash_df.rename(columns={0:hash_col},inplace=True)

    # create a geometry column to store the geometry
    all_geohash_df['geometry'] = None
    
    # populate the geometry column with shapely objects of the respective geohash
    all_geohash_df.geometry = all_geohash_df.apply(lambda x:geohash_to_polygon(x[hash_col]),axis=1)
    
    # set the geometry column of the geodataframe
    all_geohash_df = all_geohash_df.set_geometry('geometry')
    
    # set the crs
    all_geohash_df.crs = {'init' :'epsg:4326'}
    
    return all_geohash_df
```

```python
sf_all_geohash_df = get_geohash_from_bounding_box(sf_gdf_bb)

# how many hashes do we have?
print(len(sf_all_geohash_df))
sf_all_geohash_df.head()
```

### H3 

https://uber.github.io/h3/#/documentation/core-library/resolution-table

```python
def get_h3_from_bounding_box(gdf):
    # select the hex size to work
    APERTURE_SIZE = 10
    hex_col = 'hex' + str(APERTURE_SIZE)
    
    gdf_h3_list = list(h3.polyfill(gdf.geometry[0].__geo_interface__, APERTURE_SIZE, geo_json_conformant=True))
    
    # create an empty dataframe with all the hex
    all_h3_df = pd.DataFrame(gdf_h3_list)
    all_h3_df.rename(columns={0:hex_col},inplace=True)

    # create a geometry column to store the geometry
    all_h3_df['geometry'] = None
    
    # populate the geometry column with shapely objects of the respective geohash
    all_h3_df.geometry = all_h3_df.apply(lambda x:geometry.Polygon(h3.h3_to_geo_boundary(x[hex_col], geo_json=True)),axis=1)

    # set the geometry column of the geodataframe
    all_h3_df = all_h3_df.set_geometry('geometry')
    
    # set the crs
    all_h3_df.crs = {'init' :'epsg:4326'}
    
    return all_h3_df
```

### S2 

#sizes
http://s2geometry.io/resources/s2cell_statistics

```python
import pywraps2 as s2
```

```python
def get_s2_cell_geom_from_cellid(cellid):
    cell = s2.S2Cell(cellid)
    vertices = []
    for i in range(0, 4):
        vertex = cell.GetVertex(i)
        latlng = s2.S2LatLng(vertex)
        vertices.append([latlng.lng().degrees(),
                         latlng.lat().degrees()])
    return vertices
```

```python
def get_s2_from_bounding_box(gdf):
    
    # create a rect in s2
    region_rect = s2.S2LatLngRect(s2.S2LatLng.FromDegrees(gdf.total_bounds[1],gdf.total_bounds[0])
                                  ,s2.S2LatLng.FromDegrees(gdf.total_bounds[3],gdf.total_bounds[2]))
    
    # ask s2 to create a cover of this rect
    coverer = s2.S2RegionCoverer()
    coverer.set_fixed_level(16)
#     coverer.set_min_level(10)
#     coverer.set_max_level(30)
#     coverer.set_max_cells(60)
    covering = coverer.GetCovering(region_rect)

    gdf_s2_list = covering
    gdf_s2_token = [c.ToToken() for c in covering]

    # create an empty dataframe with all the hex
    all_s2_df = pd.DataFrame(gdf_s2_list)
    all_s2_df.rename(columns={0:'s2'},inplace=True)

    all_s2_df['s2Token'] = gdf_s2_token


    # create a geometry column to store the geometry
    all_s2_df['geometry'] = None

    # populate the geometry column with shapely objects of the respective geohash
    all_s2_df.geometry = all_s2_df.apply(lambda x:geometry.Polygon(get_s2_cell_geom_from_cellid(x['s2'])),axis=1)

    # set the geometry column of the geodataframe
    all_s2_df = all_s2_df.set_geometry('geometry')
    
    # set the crs
    all_s2_df.crs = {'init' :'epsg:4326'}

    return all_s2_df
```

## Calculate areas, lengths and shape distortion of each hash

some libraries may include functions to calculate areas or length etc. h3 and s2 have

```python
# def get_max_distance_centroid_to_vertex(geom):

#     centroid_df = gpd.GeoDataFrame(geom.centroid).set_geometry(0)
#     centroid = centroid_df.geometry.iloc[0]
#     max_dist = centroid.hausdorff_distance(geom.geometry.iloc[0])
#     return max_dist

def get_max_distance_centroid_to_vertex(geom):

    centroid = geom.centroid
    max_dist = centroid.hausdorff_distance(geom.exterior)
    return max_dist

def get_min_distance_centroid_to_vertex(geom):

    centroid = geom.centroid
    min_dist = centroid.distance(geom.exterior)
    return min_dist

def get_area_in_sq_km(geom):
    """
    Compute the area in km^2 of a shapely geometry, whose points are in longitude and latitude.

    Parameters
    ----------
    geometry: shapely geometry
        Points must be in longitude and latitude.

    Returns
    -------
    area:  float
        Area in m^2 due to the small size of the cell

    """
    
    import pyproj
    from shapely.ops import transform
    from functools import partial

    geom_aea = transform(
    partial(
        pyproj.transform,
        pyproj.Proj(init='EPSG:4326'),
        pyproj.Proj(
            proj='aea',
            lat_1=geom.bounds[1],
            lat_2=geom.bounds[3])),
    geom)
    # for area in km2 #/1e6
    return geom_aea.area, geom_aea.length,\
        get_max_distance_centroid_to_vertex(geom_aea),\
        get_min_distance_centroid_to_vertex(geom_aea)  

def run_area_calculation_on_df(df):
    df['geom_area']  = df.apply(lambda x :get_area_in_sq_km(x.geometry)[0], axis=1)
    df['geom_length'] = df.apply(lambda x :get_area_in_sq_km(x.geometry)[1], axis=1)
    df['max_dist'] = df.apply(lambda x :get_area_in_sq_km(x.geometry)[2], axis=1)
    df['min_dist'] = df.apply(lambda x :get_area_in_sq_km(x.geometry)[3], axis=1)
    return df

def add_area_column_on_df(df):
    
    import multiprocessing
    from multiprocessing import cpu_count
    import numpy as np
    nCores = cpu_count()
    
    data_split = np.array_split(df, nCores)
    
    with multiprocessing.Pool(processes=nCores) as pool:
        new_df = pd.concat(pool.map(run_area_calculation_on_df,data_split))
        
    return new_df
```
